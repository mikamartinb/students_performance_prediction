# Import requirements

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Read data

```python
path = "../data/StudentsPerformance.csv"
train_data = pd.read_csv(path)
train_data
```

# One-Hot-Encoding

```python
train_data["test preparation course"].replace({"none": 0, "completed": 1}, inplace=True)
train_data["lunch"].replace({"standard": 0, "free/reduced": 1}, inplace=True)
train_data["gender"].replace({"female": 0, "male": 1}, inplace=True)
train_data["race/ethnicity"].replace({"group A": 0, "group B": 1, "group C": 2 ,"group D": 3, "group E": 4}, inplace=True)
train_data["parental level of education"].replace({"some college": 0, "bachelor's degree": 1, "master's degree": 2 ,"associate's degree": 3, "high school": 4, "some high school": 5}, inplace=True)

train_data
```

# Save processed data

```python
train_data.to_csv("../data/Processed_StudentsPerformance.csv", index=False)
```
